```python
import pandas as pd
```

```python
iris_data = pd.read_csv('iris.csv')
X = iris_data[['sepal.length', 'sepal.width', 'petal.length', 'petal.width']]
y = iris_data[['variety']]
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X_pca[y == 0, 0], X_pca[y == 0, 1])
plt.scatter(X_pca[y == 1, 0], X_pca[y == 1, 1])
plt.scatter(X_pca[y == 2, 0], X_pca[y == 1, 1])
plt.show()
```

```python
from sklearn.manifold import TSNE

tsne  = TSNE(n_components=2)
X_tsne = tsne .fit_transform(X)
```

```python
plt.scatter(X_tsne[y == 0, 0], X_tsne[y == 0, 1])
plt.scatter(X_tsne[y == 1, 0], X_tsne[y == 1, 1])
plt.scatter(X_tsne[y == 2, 0], X_tsne[y == 2, 1])
plt.show()
```
